```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns= 'User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace = True)
```

```python
pt.head(30)
```

Now calculate distances to make proper predictions

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score= cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    #first fetch the index of the given book from this function
    x = np.where(pt.index == book_name)[0][0]
    distances = list(enumerate(similarity_score[x]))
    z = sorted(distances, key = lambda x:x[1])
    z= z[::-1]
    similar_items= z[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data

  
    
    
```

```python
recommend('We Were the Mulvaneys' )
```
